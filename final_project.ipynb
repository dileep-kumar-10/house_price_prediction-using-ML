```python
def clean(sqft):
    tokens=sqft.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
clean('200')
```

```python
data["total_sqft"]=data["total_sqft"].apply(clean)
data["total_sqft"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

```

```python
col_trans=ColumnTransformer((OneHotEncoder(sparse_output=False),["location"]),remainder="passthrough")
```

```python
lr=LinearRegression()
```

```python
scaler=StandardScaler()
```

```python
data3
```

```python
x = data3[['location', 'total_sqft', 'bath', 'balcony', 'bedrooms']]
y = data3['price']
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=42)
numeric_cols = ['total_sqft', 'bath', 'balcony', 'bedrooms']
categorical_cols = ['location']
preprocess = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)])
lr = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', LinearRegression())])
lr.fit(x_train, y_train)
print("Linear Regression R² Score:", lr.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', RandomForestRegressor(n_estimators=200, random_state=42))])
rf.fit(x_train, y_train)
print("Random Forest R² Score:", rf.score(x_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', DecisionTreeRegressor(random_state=42))])
dt.fit(x_train, y_train)
r2_score = dt.score(x_test, y_test)
print("Decision Tree Regressor R² Score:", r2_score)
```

```python
final_data=pd.DataFrame({"models":["LR","RF","DTR"],"r_score":[lr.score(x_test, y_test),rf.score(x_test, y_test),r2_score]})
final_data

```
